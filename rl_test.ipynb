```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("weather.csv")
```

```python
alpha = 0.1
gamma = 0.9
target = 22
epsilon = 0.5
decary = 0.8
episodes = 500
```

```python
n_states = int(data["Temp_C"].max() - data["Temp_C"].min())
target-= data["Temp_C"].min()
data["Temp_C"] -= data["Temp_C"].min()
target = int(target)
actions = [0 , 1 , -1]
q_table = np.zeros((n_states , 3))
```

```python
target = 3
```

```python
temps = data["Temp_C"]
```

```python
def get_reward(state , action):
    r = -abs(target - state)
    if(action!=1):
        r-= 0.2
    return r
for _ in range(episodes):
    temp = temps[0]
    heat = 0
    for i in range(len(temps) -1 ):
        state = int(temp)
        val = np.random.rand()
        action = None
        if(state==target): break
        state = int(max(0 , min(n_states-1 , temp)))
        if(val<=epsilon):
            action = np.random.randint(3)
        else:
            action = q_table[state].argmax()
        epsilon *= 0.995
        reward = get_reward(state , action)
        heat = heat*decary  + action
        out_change = temps[i+1] - temps[i] 
        new_state = state + out_change + heat
        new_state = int(max(0 , min(n_states-1 , new_state)))
        ma = 0
        for a in actions :
            ma = max(ma , get_reward(new_state , a) + gamma*q_table[new_state][a])
        q_table[state][action] = get_reward(state , action) + alpha*(ma  - q_table[state][action])
        temp = new_state
```

```python
temp = int(temps[0])
rewards = []
heat = 0
print(target)
for i in range(len(temps)-1):
    if(temp==target): 
        break
    temp = max(0 , min(n_states - 1 , temp))
    print(i , temp)
    action = q_table[temp].argmax()
    new_state = (heat*decary + action) + state + temps[i+1] - temps[i]
    temp = new_state
    rewards.append(get_reward(temp , action))
print(np.mean(rewards))
```
